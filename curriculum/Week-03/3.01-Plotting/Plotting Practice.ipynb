```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('cereal.csv', sep=';')
```

```python

data[['fat','protein','carbo']].plot(kind='bar', figsize=(20,20))
```

```python
for column in ["fat", "protein", "carbo"]:
    sns.swarmplot(x=column, y="rating", data=data)
    sns.plt.show()
```
